```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
d = pd.read_csv('dataset_mood_smartphone.csv', index_col=0)
d['time'] = pd.to_datetime(d.time)
d['date'] = d.time.apply(lambda t: t.strftime('%Y%m%d'))
```

```python
d.head()
```

You can see that the data is very irregular, not every observation exists for each user, and the number of observations per date/per user is different as well:

```python
d.groupby(['id', 'date', 'variable']).value.count()
```

To analyse the data, we can take averages of each observation per user per date, and put them into columns:

```python
vals = d.groupby(['id', 'date']).apply(lambda g: g[['variable', 'value']].groupby('variable').mean().T)
```

There are a lot of missing values:

```python
vals
```

Let's fill them with 0's

```python
vals.fillna(0, inplace=True)
```

we can see the correlation of these averages per user:

```python
corr = vals.loc['AS14.01'].corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```

.. or for all users:

```python
corr = vals.corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```

we can do the same for max values:

```python
vals = d.groupby(['id', 'date']).apply(lambda g: g[['variable', 'value']].groupby('variable').max().T)
vals.fillna(0, inplace=True)
```

```python
corr = vals.corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```
